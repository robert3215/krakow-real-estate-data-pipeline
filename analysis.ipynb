```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 

df = pd.read_excel("data/apartments_cleaned.xlsx")
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.show()
```

```python
sns.scatterplot(data=df, x='Price', y='LivingArea', size='PricePerM2', alpha=0.6)
plt.show()
```

```python
sns.regplot(data=df, x='Price', y='LivingArea', scatter_kws={'alpha': 0.3})
plt.title('Cena vs. Metra z Lini Regresji')
plt.show()
```

```python
plt.figure(figsize=(12, 6))

sns.boxplot(data=df, x='PricePerM2', hue= 'District')
plt.show()
```

```python
sns.pairplot(data=df, vars= ['Floor','LivingArea','Price', 'Rooms'])
plt.show()
```
